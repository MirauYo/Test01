```python
import cv2
import matplotlib.pyplot as plt

# Read the grayscale image
gray_image = cv2.imread('C:\Ai_minor\AI_ComputerVision\dataset\colorobject.png', cv2.IMREAD_GRAYSCALE)

# Calculate histogram
hist = cv2.calcHist([gray_image], [0], None, [256], [0, 256])

# Plot histogram
plt.plot(hist)
plt.title('Histogram')
plt.xlabel('Pixel Intensity')
plt.ylabel('Frequency')
plt.show()
```
